```python
import rdflib
import csv
import pandas as pd
import re
import os
import json
```

```python
directory = 'relation_linking_rnliwod_results'
count = 0

#Load the appropriate json dataset for comparison
json_file = open('dataset.json')
json_data = json.load(json_file)
json_data = json_data["data"]

#read the particular csv which is seperated by "#"
data = pd.read_csv("test_csv_relation_linking.csv",sep='#')

final_answer = []
for filename in os.listdir(directory):
    if(filename.endswith(".ttl")):#Read all the .ttl files in the directory
        count = count + 1
        file_path = os.path.join(directory, filename)
        print "file_path =>",file_path
        question_number = file_path.split("/")[1].split(".")[0].split("_")[1]
        question_number = int(question_number)
        print "number =>",question_number
        print "Q =>",data['col2'][question_number].encode("utf-8")
        gold_standard = []
        for i in range(len(json_data)):
            if(data['col2'][question_number].encode("utf-8") == json_data[i]['question'].encode("utf-8")):
                g_s = json_data[i]['predicate mapping']
                for j in range(len(g_s)):
                    uri = json_data[i]['predicate mapping'][j]['uri']
                    name = uri.split("/")[-2]
                    value = uri.split("/")[-1]
                    if(name == "property"):
                        gold_standard.append(value.encode("utf-8"))
                    else:
                        if(value[0] == value[0].lower()):
                            gold_standard.append(value.encode("utf-8"))
        g=rdflib.Graph()
        g.load(file_path,format="ttl")
        derived = []
        for s,p,o in g:
            #print "s =>",s
            #print "p =>",p
            #print "o =>",o
            a = re.findall('https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,}',o)
            if(len(a) > 0):
                url_name = a[0].split("/")[2]
                #print "URL Name= > ",url_name
                
                if(url_name == "dbpedia.org"):
                    print "url_name =>",a
                    relation_type = a[0].split("/")[-2]
                    relation_name = a[0].split("/")[-1]
                    #print "type =>",relation_type
                    #print "name =>",relation_name
                    if(relation_name[0] == relation_name[0].lower()):
                        derived.append(relation_name.encode("utf-8"))
                    #print "URL entity Type => ",a[0].split("/")[3]
                    #print "URL entity Name => ",a[0].split("/")[4]
                    #derived.append(a[0].split("/")[4].encode("utf-8"))
        print "DERIVED =>",derived
        print "GOLD    =>",gold_standard
        correct_entities = 0
        for i in derived:
            for j in gold_standard:
                if (i == j):
                    correct_entities = correct_entities + 1
        if correct_entities == 0:
            precision = 0
            recall = 0
            f_score = 0
        else:
            precision = correct_entities/float(len(derived))
            recall = correct_entities/float(len(gold_standard))
            f_score = (2*precision*recall)/(precision+recall)
        print "CORRECT =>",correct_entities
        print "precision =>",precision
        print "recall =>",recall
        print "f_score =>",f_score
        answer = []
        answer.append(data['col2'][question_number].encode("utf-8"))
        answer.append(precision)
        answer.append(recall)
        answer.append(f_score)
        answer.append(derived)
        answer.append(gold_standard)
        final_answer.append(answer)
        print "==================================================================================="
```

```python
columns = ['question','precision','recall','f_score','derived','gold_standard']
df = pd.DataFrame(final_answer,columns = columns)
writer = pd.ExcelWriter('relation_linking_score_rnliwod.xlsx',engine='openpyxl')
df.to_excel(writer,'Sheet1')
writer.save()
```
